## Tuplas

Las tuplas son estructuras inmutables que almacenan valores.

Hay diferentes formas de asignar tuplas:

```python
tup = ()

tup2 = tuple()

tup3 = tuple([1,2,3])

tup4 = (1,2)
```

```python
type(tup), type(tup2), type(tup3), type(tup4) # notar que el resultado de esto también es una tupla
```

```python
tuple1 = ('fantasia',10,5)
```

## List

Una lista es una secuencia de elementos. La posición de un elemento adentro de lal ista es denominada ``index``, los cuales son utilizados para acceder y referirse a los elementos en ella.

image.png

Hay diferentes formas de crear una lista:

```python
list1 = []

list2 = list()

list3 = ['a','b','c']

listi = ['Harry Potter y la Piedra Filosofal',2001,5]
```

```python
list1, list2, list3
```

También podemos usar índices positivos y negativos:

image.png

```python
print(listi[0], '--', listi[-3])

print(listi[1], '--', listi[-2])

print(listi[2], '--', listi[-1])
```

Las listas pueden contener cualquier tipo de elementos, incluyendo otras listas o tuplas.

```python
['Harry Potter y la Piedra Filosofal',2001,5,['Ron','Harry'],('Hermione','gato')]
```

### Operaciones 

```python
listi = ['Piedra Filosofal', 'Cámara de los Secretos', 'Prisionero de Azcaban', 'Caliz de Fuego', 'Orden del Fenix', 'Principe Mestizo', 'Reliquias de la Muerte']
```

#### Slicing

Al igual que hicimos con los strings, podemos obtener slices de la lista.

```python
# seleccionamos los elementos 3 y 4
listi[3:5]
```

#### extend

El ``addAll`` para listas. 

```python
listi.extend(['Cursed Child', 'Hall of Elders\' Crossing'])
listi
```

#### append

Agrega un elemento al final de la lista. No importa qué tipo tiene el elemento a agregar.

```python
listi.append(['Cursed Child', 'Hall of Elders\' Crossing']) # esto agrega la lista como elemento, generando una lista anidada
listi
```

#### insert 

Inserta un elemento ``y`` en la posición especificada ``x``. No reemplaza el elemento que previamente se encontraba en dicha poisición, sino que los desplaza. Para reemplazar, simplemente hacer una asignación.

```python
listi.insert(0,'Fantastic Beasts')

listi
```

Las listas son mutables, con lo que los elementos pueden ser modificados.

```python
A = ['Harry Potter y la Piedra Filosofal',2001,5]

print(A)

A[0] = 'El señor de los anillos' # cambiamos el primer elemento de lalista

print(A)
```

#### pop

Obtener el último elemento en la lista, o el tope de una pila. Al obtener el elemento lo remueve de la lista.

Se le puede pasar índice para determinar qué valor obtener.

```python
listi = [1,2,3,4]

print(listi.pop())

print(listi)
```

#### del

Para borrar un elemento. No es una función de la lista, sino que hay que invocarlo con el elemento a eliminar.


```python
print(A)

del(A[0])

print(A)
```

#### remove

Alternativa al ``del``. Esta si es una función de la lista. Se le debe pasar el elemento a eliminar. Borra una única aparición del elemento.

```python
listi = [1,2,3,4,4]

listi.remove(4)

listi
```

#### reverse

Invertir el orden de la lista.

```python
listi = [1,2,3,4]

listi.reverse()

listi
```

En el caso de que la lista contenga otra lista anidada, la misma no es invertida.

```python
listi = [1,2,3,[5,6,7],4]

listi.reverse()

listi
```

#### sort

Ordena los elementos de la lista en orden natural ascendente. Se le puede definir ``reverse=True`` para que el orden sea descendente.

En este caso, si hubiese listas anidadas, da error.

```python
listi = [1, 1, 2, 7, 5, 7, 8]

listi.sort()

listi
```

```python
listi.sort(reverse=True)

listi
```

#### split

Convierte un string en una lista. Particiona el string sobre el cual se invoca de acuerdo al char que se pase por parámetro. Por defecto, particiona sobre los espacios en blanco.

```python
'harry potter y la piedra filosofal'.split()
```

```python
'harry potter y la piedra filosofal'.split('y')
```

#### Copia vs clonación

Dada una lista ``A`` si la asignamos a ``B``, tanto ``A`` como ``B`` estarán referenciando a la misma lista en memoria.

```python
A = ['Harry','Potter']

B = A

print(A)
print(B)
```

Esto quiere decir que si hacemos cambios en una lista, los mismos se verán reflejados en la otra.

```python
print(B[0])

A[0] = "banana"

print(B[0])
```

```python
Para clonar una lista:
```

```python
B = A[:]

B
```

Ahora, cambios en ``A`` no afectarán a ``B``.

```python
print(B[0])

A[0] = "Harry"

print(B[0])
```

## Sets

Un set es una colección de objtos únicos, representa el concepto matemático de conjunto. Se lo señala con llaves ``{}``.

En caso de que al construir se pasen elementos repetidos, Python los eliminará de forma automática.

```python
seti = {"fantasy", 'horror', 'fantasy', 'crime', 'noir'}
seti
```

También se puede construir un set a partir de una lista:

```python
listi = ["fantasy", 'horror', 'fantasy', 'crime', 'noir']
seti = set(listi)   

seti
```

### Operaciones

#### add

Agregar un elemento al set.

```python
A = {'crime', 'noir'}

A.add('thriller')

A
```

Si agregamos nuevamente el elemento, no será agregado dado que ya existe.

```python
A.add('thriller')

A
```

#### remove

Eliminar un elemento del set. Eliminar un elemento que no se encuentra en el set resulta en un error.

```python
A.remove('noir')

A
```

#### in

Verificar si un elemento se encuentra en el set.

```python
'crime' in A
```

```python
'noir' in A
```

#### clear

Vaciar el set.

```python
A.clear()

A
```

### Operaciones lógicas

Con los sets se pueden hacer operaciones como diferencia, diferencia simétrica, intersección y unión.

```python
# sets a utilizar en los ejemplos

set1 = {"fantasy", 'crime', 'noir'}
set2 = {"crime", "noir", "scifi"}
```

image.png

#### intersección

La intersección (los elementos que se encuentran en ambos sets) se puede encontrar usando ``&``.

image.png

```python
set1 & set2
```

#### diferencia

Permite encontrar los elementos que están en un set (sobre el que se invoca el método), pero no en el otro (pasado por parámetro).

image.png

```python
set1.difference(set2) 
```

image.png

```python
set2.difference(set1) 
```

#### union

Encontrar los elementos en que se encuentran en al menos uno de los sets.

image.png

```python
set1.union(set2)
```

#### superset

Determinar si un set contiene todos los elementos del otro set.

```python
set1.issuperset(set2)
```

```python
set11 = {1,2,3}
set22 = {1}

set11.issuperset(set22)
```

#### subset

Determinar si un set se encuentra completamente contenido en otro. Es decir, si todos los elementos de un set se encuentran en el otro.

```python
set1.issubset(set2)
```

```python
set11 = {1}
set22 = {1,2,3}

set11.issubset(set22)
```

#### disjoint

Determinar si los sets no tienen elementos en común.

```python
set1.isdisjoint(set2)
```

## Diccionarios/mapas

Un diccionario o mapa consiste de claves y valores. Puede ser considerada como una lista, que en lugar de tener índices numéricos tiene índices por clave, las cuales son utilizadas para acceder a los valores en el diccionario.

Las claves deben ser elementos inmutables.

image.png

Los diccionarios pueden crearse vacíos y luego agregar elementos haciendo asignaciones:

```python
dicti1 = dict()

dicti1['key1'] = 1

dicti1
```

O pueden crearse con valores iniciales:

```python
dicti1 = {'key1' : 1, 'key2' : 2}

dicti1
```

Para acceder a los valores, se puede usar la sintaxis de indexado como para las tuplas y listas:

```python
dicti1['key1']
```

Sin embargo, esto dará error si la key no existe:

```python
dicti1['key3']
```

En estos casos, se puede hacer una verificación de si el valor existe en el diccionario con un ``in`` (al igual que en los sets), o utilizar la función ``get`` que retorna ``None`` por defecto si la clave no está en el diccionario. 

El ``get`` permite setear el valor por defecto a retornar.

```python
print(dicti1.get('key3'))
```

### del 

Al igual que para las estructuras anteriores, se puede utilizar ``del`` para eliminar una key (y su correspondiente valor).

```python
del(dicti1['key1'])

dicti1
```

#### keys

Permite obtener las keys en el diccionario.

```python
dicti1.keys()
```

#### values

Permite obtener los valores en el diccionario.

```python
dicti1.values()
```

#### items

Permite obtener los pares de ``(key, valor)`` en el diccionario.

```python
dicti1.items()
```
